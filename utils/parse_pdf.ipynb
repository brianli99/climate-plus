```python
import pandas as pd
from tqdm import tqdm
import requests
import PyPDF2

df = pd.read_csv("../data/tcfd.csv")
```

```python
i = 0

missed_files = []

for i in tqdm(range(len(df['Report URL']))):
    # print("Downloading file: ", i)
    try:
        response = requests.get(df['Report URL'][i], timeout=5)
        pdf = open("../data/pdf/"+str(i)+".pdf", "wb")
        pdf.write(response.content)
        pdf.close()
        # print("File ", i, " downloaded")
    except:
        print("File ", i, " NOT downloaded")
        missed_files.append(i)
    i += 1
print("All files downloaded")
```

```python
def parse_pdf_from_path(file_path):
    # Open the PDF file
    with open(file_path, 'rb') as f:
        # Create a PDF reader object
        pdf_reader = PyPDF2.PdfReader(f)

        # Extract the text from each page
        text = ''
        for i in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[i]
            text += page.extract_text()

        return text
```

```python
def write_to_txt_file(text, file_path):
    with open(file_path, 'w') as f:
        f.write(text)
```

```python
for i in tqdm(range(len(df['Report URL']))):
    try:
        text = parse_pdf_from_path("../data/pdf/"+str(i)+".pdf")
        write_to_txt_file(text, "../data/txt/"+str(i)+".txt")
    except:
        print("File ", i, " NOT parsed")
    i += 1
```
